```python
import numpy as np
import tensorflow as tf
```

```python
corpus_raw = 'He is the king . The king is royal . She is the royal  queen '
```

```python
corpus_raw = corpus_raw.lower()
```

```python
words = []

for word in corpus_raw.split():
    if word != ".":
        words.append(word)

# Usuwanie duplikatów
words = set(words)
```

```python
word2int = {}
int2word = {}
vocab_size = len(words)
```

```python
for i, word in enumerate(words):
    word2int[word] = i
    int2word[i] = word
```

```python
word2int['queen']
```

```python
raw_sentences = corpus_raw.split('.')
sentences= []
```

```python
for sentence in raw_sentences:
    sentences.append(sentence.split())
```

```python
sentences
```

```python
data = []
WINDOW_SIZE = 2

for sentence in sentences:
    for word_index, word in enumerate(sentence):
        for nb_word in sentence[max(word_index - WINDOW_SIZE, 0):
min(word_index + WINDOW_SIZE, len(sentence))+1]:
            if nb_word != word:
                data.append([word, nb_word])
```

```python
data
```

```python
def to_one_hot(data_point_index, vocab_size):
    temp = np.zeros(vocab_size)
    temp[data_point_index] = 1
    return temp
```

```python
x_train = []
y_train = []
```

```python
for data_word in data:
    x_train.append(to_one_hot(word2int[data_word[0]], vocab_size))
    y_train.append(to_one_hot(word2int[ data_word[1] ], vocab_size))
```

```python
x_train = np.asarray(x_train)
y_train = np.asarray(y_train)
```

```python
word2int
```

```python
x = tf.placeholder(tf.float32, shape=(None, vocab_size))
y_label =tf.placeholder(tf.float32, shape=(None, vocab_size))
```

```python
# Network
model = tf.layers.dense(x, 2)
output = tf.nn.softmax(tf.layers.dense(model, 7))
```

```python
# Loss function
loss = tf.reduce_mean(-tf.reduce_sum(y_label * tf.log(output), reduction_indices=[1]))
# Clipp
loss = tf.clip_by_value(loss, 0.1, 2.)
```

```python
train_op = tf.train.GradientDescentOptimizer(.1).minimize(loss)
n_iter = 100000
```

```python
init = tf.global_variables_initializer()

sess = tf.Session()    

sess.run(init)
    
for it in range(n_iter):
    loss_val, _ = sess.run([loss, train_op], feed_dict={x: x_train, y_label: y_train})
    if it % 100 == 0:
        print "{} | loss: {}".format(it, loss_val)
```
