# Konwolucje na obrazach

```python
import numpy as np
from scipy.signal import convolve2d
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
img = mpimg.imread('lena.jpg')
```

```python
%matplotlib inline
```

```python
plt.imshow(img)
plt.show()
```

```python
img.shape
```

```python
bw = img.mean(axis=2)
plt.imshow(bw, cmap='gray')
plt.show()
```

```python
W = np.zeros((20, 20))

for i in xrange(20):
    for j in xrange(20):
        dist = (i - 9.5) ** 2 + (j - 9.5) ** 2 
        W[i, j] = np.exp(-dist / 50)
    
plt.imshow(W, cmap='gray')    
```

```python
out = convolve2d(bw, W)
plt.imshow(out)
```

```python
out = convolve2d(bw, W, mode='same')
```

```python
plt.imshow(out)
```

```python
# Operator Sobela
Hx = np.array(
[[-1, 0, 1],
[-2, 0, 2],
[-1, 0, 1]], dtype=np.float32)
```

```python
Hy = Hx.T
```

```python
Hy
```

```python
Gx = convolve2d(bw, Hx)
```

```python
plt.imshow(Gx, cmap='gray')
```

```python
Gy = convolve2d(bw, Hy)
plt.imshow(Gy, cmap='gray')
```

```python
# Wszystkie krawędzie
G = np.sqrt(Gx**2 + Gy**2)
plt.imshow(G, cmap='gray')
```

```python
theta = np.arctan2(Gy, Gx)
plt.imshow(theta, cmap='gray')
```
